```python
### converting numpy array
df1=df.values
```

```python
df1
```

```python
df1_norm=scale(df1)
```

```python
df1_norm
```

```python
### PCA
```

```python
pca=PCA()
df2_pca=pca.fit_transform(df1_norm)
```

```python
df2_pca
```

```python
pca.components_
```

```python
### amount of variance
```

```python
var=pca.explained_variance_ratio_
var
```

```python
var1=np.cumsum(np.round(var,4)*100)
```

```python
var1
```

```python
### variance plot
```

```python
plt.plot(var1,color='orange')
```

```python
fdf=pd.concat([df['Type'],pd.DataFrame(df2_pca[:,0:3],columns=['PC1','PC2','PC3'])],axis=1)
fdf
```

```python
fig=plt.figure(figsize=(16,22))
sns.scatterplot(data=fdf)
```

```python
### hirearchical cluster
```

```python
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
from sklearn.preprocessing import normalize
```

```python
plt.figure(figsize=(10,7))
dendrogram=sch.dendrogram(sch.linkage(df1_norm,'complete'))
```

```python
hclusters=AgglomerativeClustering(n_clusters=3,affinity='euclidean',linkage='ward')
hclusters
```

```python
y=pd.DataFrame(hclusters.fit_predict(df1_norm),columns=['clusterid'])
```

```python
y['clusterid'].value_counts()
```

```python
df3=df.copy()
df3['clustersid']=hclusters.labels_
df3.head()
```

```python
### K means clustering
```

```python
from sklearn.cluster import KMeans
```

```python
wcss=[]
for i in range(1,6):
    kmeans=KMeans(n_clusters=i,random_state=2)
    kmeans.fit(df1_norm)
    wcss.append(kmeans.inertia_)
```

```python
### elbow graph
```

```python
plt.plot(range(1,6),wcss)
plt.title('Elbow Graph')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
### using k=3
```

```python
cluster3=KMeans(3,random_state=30).fit(df1_norm)
cluster3
```

```python
cluster3.labels_
```

```python
df4=df.copy()
df4['clusterid']=cluster3.labels_
df4.head()
```
